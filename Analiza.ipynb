# Analiziranje Pokemonov

Analiziral bom vseh 1025 različnih pokemon in jih primerjal po tipih, potezah in po njihovi statistiki. Vse podatke sem pobral iz spletne strani Serebii

```python
import pandas as pd
import matplotlib.pyplot as plot
from Dodatne_funcije_za_analizo import *
from collections import Counter
```

Prikaže tabelo vseh pokemonov z osnovnimi podatki:

```python
pokemoni = pd.read_csv("pokemoni.csv", index_col="Stevilka")
pokemoni
```

Prikazuje enega mojih najljubših pokemonov in njegove značilnosti

```python
pokemoni.iloc[37]
```

Tabela pokemonov ketari imajo tako Hp kot tudi Attack več kot 90

```python
pokemoni[(pokemoni.Hp > 90) & (pokemoni.Attack > 90)]
```

Tabeli smo dodali stolpec ki sešteje in prikaže vse različne statiskike pokemonov in jo po tem stolpcu uredili

```python
pokemoni["Total_stats"] = pokemoni.Hp + pokemoni.Attack + pokemoni.Defense + pokemoni.Sp_Att + pokemoni.Sp_Def + pokemoni.Speed
pokemoni.sort_values(["Total_stats"])
```

Naslednja tabela prikazuje pokemone, ki imajo natanko 3 znacilne poteze

```python
pokemoni["Signature_move"] = pokemoni["Signature_move"].apply(niz_v_seznam)
pokemoni[pokemoni["Signature_move"].apply(len) == 3]
```

Prikazuje pokemone, ki imajo samo en tip

```python
pokemoni["Type"] = pokemoni["Type"].apply(niz_v_seznam)
pokemoni[pokemoni["Type"].apply(len) == 1]
```

Dodamo še en stolpec, ki pokemone razvrsti v generacije

```python
pokemoni["Generacija"] = pokemoni.index.to_series().apply(doloci_generacijo)
pokemoni
```

Graf prikazuje povprecno stevilo total stats glede na vsako generacjo

```python
average_stats_per_generation = pokemoni.groupby('Generacija')['Total_stats'].mean()
average_stats_per_generation.plot(kind='bar', color='skyblue')

plot.title('Povprecno total stats na generacijo')
plot.xlabel('Generacija')
plot.ylabel('Povprecno total stats')
plot.xticks(rotation=45)  
plot.grid(axis='y', linestyle='--', alpha=0.7)
```

Graf prikazuje število pokemonov, ki imajo nadpovprečne total stats. Če primerjamo zgornji in spodnji gradf lahko opazimo, da je nadpovprečnih pokemonov v generaciji 6 zelo malo. Kljub temu je povprečje total stats v tej generaciji dokaj visoko

```python
povprecni_total_stats = pokemoni["Total_stats"].mean()
pokemoni[pokemoni.Total_stats >= povprecni_total_stats].groupby("Generacija").size().plot(color="pink")

plot.title("Število pokemonov z nadpovprečnim total stats")
plot.ylabel("Število")
plot.grid(axis='y', linestyle='--', alpha=0.7)
```

Ločimo pokemone glede na to ali imajo samo en tip, ali 2

```python
pokemoni["Število tipov"] = pokemoni["Type"].apply(len)
stevec_tipov = pokemoni["Število tipov"].value_counts()
stevec_tipov.plot(kind='bar', color=['skyblue', 'lightgreen'])

plot.title('Število pokemonov po številu tipov')
plot.xlabel('Število tipov')
plot.ylabel('Število')
plot.xticks(ticks=[0, 1], labels=['1 Tip', '2 Tipa'], rotation=0)
plot.grid(axis='y', linestyle='--', alpha=0.7)
```

Ločimo pokemone glede na število njihovih značilnih potez

```python
pokemoni["število signature moves"] = pokemoni["Signature_move"].apply(len)
stevec_tipov = pokemoni["število signature moves"].value_counts()
stevec_tipov.plot(kind='bar', color=['lightblue', 'pink', 'purple'])

plot.title('Število pokemonov po številu potez')
plot.xlabel('Število potez')
plot.ylabel('Število')
plot.xticks(ticks=[0, 1, 2], labels=['1 Poteza', '2 Potezi', '3 Poteze'], rotation=0)
plot.grid(axis='y', linestyle='--', alpha=0.7)
```

Z grafom prikažemo top 10 "najbolših pokemonov"

```python
top_10_pokemonov = pokemoni.sort_values(["Total_stats"]).tail(10)

top_10_pokemonov.plot(
    kind='barh',
    x='Ime',
    y='Total_stats',
    color='lightgreen',
    legend=False)

plot.title("Top 10 pokemonov")
plot.xlabel("Total Stats")
plot.ylabel("Pokemon")
```

Poiščemo najpogostejši tip pokemona in v tabeli prikaženo pokemone s tem tipom

```python
pokemoni["Type"].apply(niz_v_seznam)
vsi_tipi = pd.Series([t for types in pokemoni["Type"] for t in types])

prestej_tipe = vsi_tipi.value_counts()
najpogostejsi_tip = prestej_tipe.idxmax()
print(f"Najpogostejši pokemon type je {najpogostejsi_tip}")
```

```python
water_pokemoni = pokemoni[pokemoni["Type"].apply(lambda x: "water" in x)]
water_pokemoni
```

Poiščemo najpogostejšo značilno potezo in v tabeli prikažemo pokemone s to potezo

```python
pokemoni["Signature_move"].apply(niz_v_seznam)
vsi_signature_moves = [move for sublist in pokemoni["Signature_move"] for move in sublist]

stevec_signature_moves = Counter(vsi_signature_moves)
najpogostejsi_move, stevec_najpogostejsi_move = stevec_signature_moves.most_common(1)[0]

print(f"Najpogosteje uporabljena poteza je {najpogostejsi_move}, ki jo ima {stevec_najpogostejsi_move} pokemonov")
```
